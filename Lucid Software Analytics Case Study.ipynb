# 1. Which paywalls have the best conversion rate?
Note: Conversion rate is defined as the number of people who pay because of a given paywall
divided by the number of people who hit the paywall. An important part of this task is deciding
how you will determine whether a paywall caused a user to pay or not. Also note that a user can
continue to hit paywalls after their first payment for features that are not part of their subscription
plan (Lucidpress has multiple subscription plans as shown here ).

```python
result_output.sort_values('Conversion rate',ascending = False)
```

Backup-restore, Page-Limit,  Pdf-dpi, Storage, and Analytics are the top 5 features with the highest conversion rates.

# 3. Methodology


To answer question #1, 
1. I first inner-joined the payment dataset with the paywall dataset to get the first payment date;
2. I converted datatypes from strings to datetime in relevant columns to compare the datetimes of hitting a wall with first payment date;
3. Assuming that a user chooses to upgrade because of the last paywall they hit (while an alternative could be the user considers previous paywalls and the last paywall is just the last straw), I attributed that payment to the last paywall hit;
4. While experimenting with the dataset, I discovered that some users upgraded without hitting a paywall. As a result, there are 840 cases and 36 cases wrongly attributed to Analytics and Backup-restore. These cases were excluded from the analysis afterward;
5. I grouped the dataset by paywall to get the no. of upgrade due to each paywall;
6. I found the total no. of hitting a paywall by counting the no. of timestamp under that paywall;
7. I calculated the conversion rate by dividing no. of upgrade by total no. of hit.


To answer question #2,
1. I excluded "auto-upgrade" cases when calculating revenue;
2. I grouped the dataset by paywall and summed up the First Payment Value in each category.


# 4. Recommendations for product changes


1. while Color-Management, Custom-Fonts, Downloadable-Publication, Premium Templates, and Email-Thumb are popular paywalls (with over 1,000 hits), they have relatively low conversion rates (below 0.04). These features could be potential revenue generators. I suggest marketing these features more aggressively to increase the necessity of having these features.

2. Page-Limit and Pdf-dpi are top 2 most frequently hit paywalls. A new one-time payment (e.g. 99 cents for a one-time unlimited page use) could be introduced to entice users to try these features without paying for the full amount of a package deal.

# 5. Recommendations for further analysis

1. We could track whether a subscriber chooses to further upgrade their plan and the duration for their subscription to understand the marketability of the premium features.

2. For popular features like Page-Limit, Pdf-dpi, Storage, Color-Management, Custom-Fonts, Downloadable-Publication, Premium Templates, and Email-Thumb, we could collect more data on the user profiles to better identify user personas for targeted marketing.

3. We could also analyze the time gaps between hitting a paywall and making a payment and introduce a notification feature or promotion to entice actions.
